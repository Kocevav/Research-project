```python
merged_df['article_date_v2'] = pd.to_datetime(merged_df['article_date_v2'], errors='coerce')
df_gold_prices['Date'] = pd.to_datetime(df_gold_prices['Date'], errors='coerce')

merged_df['V2Tone_num'] = merged_df['V2Tone'].apply(lambda x: float(str(x).split(',')[0]) if isinstance(x, str) else None)
merged_df['has_gold_theme'] = merged_df['Themes'].str.contains('GOLD', na=False).astype(int)

daily_sentiment = merged_df.groupby('article_date_v2').agg(
    avg_tone=('V2Tone_num', 'mean'),
    max_tone=('V2Tone_num', 'max'),
    tone_std=('V2Tone_num', 'std'),
    article_count=('V2Tone_num', 'count'),
    has_gold_theme=('has_gold_theme', 'max')

).reset_index().rename(columns={'article_date_v2': 'Date'})

df_merged = pd.merge(df_gold_prices, daily_sentiment, on='Date', how='inner')

df_merged = df_merged.sort_values(by='Date')
df_merged['Price_t-1'] = df_merged['Price'].shift(1)
df_merged['Price_change_t-1'] = df_merged['Price'] - df_merged['Price_t-1']
df_merged['MA_3'] = df_merged['Price'].rolling(window=3).mean()
df_merged['MA_7'] = df_merged['Price'].rolling(window=7).mean()
df_merged['Volatility_3d'] = df_merged['Price'].rolling(window=3).std()
df_merged['avg_tone_ma3'] = df_merged['avg_tone'].rolling(window=3).mean()
df_merged['avg_tone_ma7'] = df_merged['avg_tone'].rolling(window=7).mean()

future_price = df_merged['Price'].shift(-3)
growth = (future_price - df_merged['Price']) / df_merged['Price']
df_merged['Target_up_0_3pct_3d'] = (growth > 0.003).astype(int)

df_merged = df_merged.dropna()
```

```python
features = [
    'Price', 'Open', 'High', 'Low', 'Change %',
    'avg_tone', 'max_tone', 'tone_std', 'article_count', 'has_gold_theme',
    'Price_t-1', 'Price_change_t-1', 'MA_3', 'MA_7', 'Volatility_3d',
    'avg_tone_ma3', 'avg_tone_ma7'
]

X = df_merged[features]
y = df_merged['Target_up_0_3pct_3d']
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import f1_score, accuracy_score
from imblearn.over_sampling import RandomOverSampler


results = []

window_size = 200
step = 50

for start in range(0, len(X) - window_size - step, step):
    end = start + window_size
    X_train = X.iloc[start:end]
    y_train = y.iloc[start:end]
    X_test = X.iloc[end:end+step]
    y_test = y.iloc[end:end+step]

    ros = RandomOverSampler(random_state=42)
    X_resampled, y_resampled = ros.fit_resample(X_train, y_train)

    model = XGBClassifier(
        n_estimators=150,
        max_depth=4,
        learning_rate=0.1,
        subsample=0.8,
        colsample_bytree=0.8,
        use_label_encoder=False,
        eval_metric='logloss'
    )
    model.fit(X_resampled, y_resampled)

    y_pred = model.predict(X_test)
    f1 = f1_score(y_test, y_pred, zero_division=0)
    acc = accuracy_score(y_test, y_pred)

    results.append({'Start': start, 'F1-score': f1, 'Accuracy': acc})
```

```python
import matplotlib.pyplot as plt

df_results = pd.DataFrame(results)

print("Средна Accuracy:", df_results['Accuracy'].mean())
print("Средна F1-score:", df_results['F1-score'].mean())


plt.figure(figsize=(10, 5))
plt.plot(df_results['Start'], df_results['F1-score'], label='F1-score', marker='o')
plt.plot(df_results['Start'], df_results['Accuracy'], label='Accuracy', marker='x')
plt.title('Rolling Evaluation (XGBoost)')
plt.xlabel('Start Index')
plt.ylabel('Score')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from xgboost import XGBClassifier
from sklearn.metrics import f1_score
from imblearn.over_sampling import RandomOverSampler
from sklearn.impute import SimpleImputer
from collections import Counter

def add_indicators(df):
    delta = df['Price'].diff()

    df['momentum'] = df['Price'] - df['Price'].shift(5)

    gain = np.where(delta > 0, delta, 0)
    loss = np.where(delta < 0, -delta, 0)
    avg_gain = pd.Series(gain).rolling(window=14).mean()
    avg_loss = pd.Series(loss).rolling(window=14).mean()
    rs = avg_gain / avg_loss
    df['RSI'] = 100 - (100 / (1 + rs))

    ema_12 = df['Price'].ewm(span=12, adjust=False).mean()
    ema_26 = df['Price'].ewm(span=26, adjust=False).mean()
    df['MACD'] = ema_12 - ema_26
    df['MACD_signal'] = df['MACD'].ewm(span=9, adjust=False).mean()

    return df

df_merged = add_indicators(df_merged)

def create_hold_focused_target(df, hold_multiplier=2.5):
    df['rolling_volatility'] = df['Price'].rolling(window=21).std() / df['Price'].rolling(window=21).mean()
    df['rolling_volatility'] = df['rolling_volatility'].fillna(df['rolling_volatility'].mean())

    future_price = df['Price'].shift(-3)
    growth = (future_price - df['Price']) / df['Price']
    targets = []

    for i in range(len(df)):
        current_growth = growth.iloc[i]
        volatility = df['rolling_volatility'].iloc[i]
        threshold = max(0.0015, volatility * 0.3)
        hold_threshold = threshold * hold_multiplier

        if current_growth > hold_threshold:
            targets.append(2)  # BUY
        elif current_growth < -hold_threshold:
            targets.append(0)  # SELL
        else:
            targets.append(1)  # HOLD

    return targets

def get_sample_weights(y):
    weights = np.ones(len(y))
    weights[y == 1] = 2.5
    return weights

def hold_focused_sampling(X, y, hold_boost=2.5):
    imputer = SimpleImputer(strategy='median')
    X_filled = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)
    class_counts = Counter(y)
    max_count = max(class_counts.values())
    sampling_strategy = {0: max_count, 1: int(max_count * hold_boost), 2: max_count}
    ros = RandomOverSampler(sampling_strategy=sampling_strategy, random_state=42)
    X_resampled, y_resampled = ros.fit_resample(X_filled, y)
    return X_resampled, y_resampled

def evaluate_hold_strategy(X, y):
    results = []
    window_size = 200
    step = 50

    model = XGBClassifier(
        objective='multi:softprob',
        num_class=3,
        eval_metric='mlogloss',
        n_estimators=120,
        max_depth=5,
        learning_rate=0.1,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42
    )

    for start in range(0, len(X) - window_size - step, step):
        end = start + window_size
        X_train, y_train = X.iloc[start:end], y.iloc[start:end]
        X_test, y_test = X.iloc[end:end+step], y.iloc[end:end+step]

        if len(set(y_train)) < 3 or len(set(y_test)) < 2:
            continue

        X_resampled, y_resampled = hold_focused_sampling(X_train, y_train)
        weights = get_sample_weights(y_resampled)
        model.fit(X_resampled, y_resampled, sample_weight=weights)
        y_pred = model.predict(X_test)

        f1_macro = f1_score(y_test, y_pred, average='macro', zero_division=0)
        f1_all = f1_score(y_test, y_pred, average=None, zero_division=0)
        results.append({
            'Start': start,
            'F1_macro': f1_macro,
            'F1_BUY': f1_all[2] if 2 in y_test.values else 0,
            'F1_HOLD': f1_all[1] if 1 in y_test.values else 0,
            'F1_SELL': f1_all[0] if 0 in y_test.values else 0
        })

    return pd.DataFrame(results)

df_merged['Target_hold_focused'] = create_hold_focused_target(df_merged, hold_multiplier=2.5)
features = ['Price', 'Open', 'High', 'Low', 'Change %', 'avg_tone', 'max_tone', 'tone_std',
            'article_count', 'has_gold_theme', 'Price_t-1', 'Price_change_t-1', 'MA_3', 'MA_7',
            'Volatility_3d', 'avg_tone_ma3', 'avg_tone_ma7', 'momentum', 'RSI', 'MACD', 'MACD_signal']

df_clean = df_merged.dropna(subset=features + ['Target_hold_focused'])
X = df_clean[features].fillna(df_clean[features].median())
y = df_clean['Target_hold_focused']

df_results = evaluate_hold_strategy(X, y)

plt.figure(figsize=(10, 5))
plt.plot(df_results['Start'], df_results['F1_BUY'], label='F1 BUY')
plt.plot(df_results['Start'], df_results['F1_HOLD'], label='F1 HOLD')
plt.plot(df_results['Start'], df_results['F1_SELL'], label='F1 SELL')
plt.plot(df_results['Start'], df_results['F1_macro'], label='Macro F1', linestyle='--')
plt.xlabel("Start Index")
plt.ylabel("F1-score")
plt.title("Rolling F1-scores (Multiclass)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print("Средна Macro F1:", df_results['F1_macro'].mean())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from xgboost import XGBClassifier
from sklearn.metrics import f1_score, accuracy_score
from imblearn.over_sampling import RandomOverSampler
from sklearn.impute import SimpleImputer
from collections import Counter

def add_indicators(df):
    delta = df['Price'].diff()
    df['momentum'] = df['Price'] - df['Price'].shift(5)
    gain = np.where(delta > 0, delta, 0)
    loss = np.where(delta < 0, -delta, 0)
    avg_gain = pd.Series(gain).rolling(window=14).mean()
    avg_loss = pd.Series(loss).rolling(window=14).mean()
    rs = avg_gain / avg_loss
    df['RSI'] = 100 - (100 / (1 + rs))
    ema_12 = df['Price'].ewm(span=12, adjust=False).mean()
    ema_26 = df['Price'].ewm(span=26, adjust=False).mean()
    df['MACD'] = ema_12 - ema_26
    df['MACD_signal'] = df['MACD'].ewm(span=9, adjust=False).mean()
    return df

def create_hold_focused_target(df, hold_multiplier=2.5):
    df['rolling_volatility'] = df['Price'].rolling(window=21).std() / df['Price'].rolling(window=21).mean()
    df['rolling_volatility'] = df['rolling_volatility'].fillna(df['rolling_volatility'].mean())
    future_price = df['Price'].shift(-3)
    growth = (future_price - df['Price']) / df['Price']
    targets = []
    for i in range(len(df)):
        current_growth = growth.iloc[i]
        volatility = df['rolling_volatility'].iloc[i]
        threshold = max(0.0015, volatility * 0.3)
        hold_threshold = threshold * hold_multiplier
        if current_growth > hold_threshold:
            targets.append(2)  # BUY
        elif current_growth < -hold_threshold:
            targets.append(0)  # SELL
        else:
            targets.append(1)  # HOLD
    return targets

def get_sample_weights(y):
    weights = np.ones(len(y))
    weights[y == 1] = 2.5
    return weights

def hold_focused_sampling(X, y, hold_boost=2.5):
    imputer = SimpleImputer(strategy='median')
    X_filled = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)
    class_counts = Counter(y)
    max_count = max(class_counts.values())
    sampling_strategy = {0: max_count, 1: int(max_count * hold_boost), 2: max_count}
    ros = RandomOverSampler(sampling_strategy=sampling_strategy, random_state=42)
    return ros.fit_resample(X_filled, y)

def evaluate_hold_strategy(X, y):
    results = []
    window_size = 200
    step = 50

    model = XGBClassifier(
        objective='multi:softprob',
        num_class=3,
        eval_metric='mlogloss',
        n_estimators=120,
        max_depth=5,
        learning_rate=0.1,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42
    )

    for start in range(0, len(X) - window_size - step, step):
        end = start + window_size
        X_train, y_train = X.iloc[start:end], y.iloc[start:end]
        X_test, y_test = X.iloc[end:end+step], y.iloc[end:end+step]

        if len(set(y_train)) < 3 or len(set(y_test)) < 2:
            continue

        X_resampled, y_resampled = hold_focused_sampling(X_train, y_train)
        weights = get_sample_weights(y_resampled)
        model.fit(X_resampled, y_resampled, sample_weight=weights)
        y_pred = model.predict(X_test)

        f1_macro = f1_score(y_test, y_pred, average='macro', zero_division=0)
        f1_all = f1_score(y_test, y_pred, average=None, zero_division=0)

        acc_buy = accuracy_score(y_test[y_test == 2], y_pred[y_test == 2]) if (y_test == 2).sum() > 0 else np.nan
        acc_hold = accuracy_score(y_test[y_test == 1], y_pred[y_test == 1]) if (y_test == 1).sum() > 0 else np.nan
        acc_sell = accuracy_score(y_test[y_test == 0], y_pred[y_test == 0]) if (y_test == 0).sum() > 0 else np.nan

        results.append({
            'Start': start,
            'F1_macro': f1_macro,
            'F1_BUY': f1_all[2] if 2 in y_test.values else 0,
            'F1_HOLD': f1_all[1] if 1 in y_test.values else 0,
            'F1_SELL': f1_all[0] if 0 in y_test.values else 0,
            'Acc_BUY': acc_buy,
            'Acc_HOLD': acc_hold,
            'Acc_SELL': acc_sell
        })

    return pd.DataFrame(results)

def plot_f1_and_accuracy(df_results):
    plt.figure(figsize=(12, 6))
    plt.plot(df_results['Start'], df_results['F1_BUY'], label='F1 BUY')
    plt.plot(df_results['Start'], df_results['F1_HOLD'], label='F1 HOLD')
    plt.plot(df_results['Start'], df_results['F1_SELL'], label='F1 SELL')
    plt.plot(df_results['Start'], df_results['F1_macro'], label='Macro F1', linestyle='--')
    plt.plot(df_results['Start'], df_results['Acc_BUY'], label='Acc BUY', linestyle=':', marker='o')
    plt.plot(df_results['Start'], df_results['Acc_HOLD'], label='Acc HOLD', linestyle=':', marker='x')
    plt.plot(df_results['Start'], df_results['Acc_SELL'], label='Acc SELL', linestyle=':', marker='s')
    plt.xlabel("Start Index")
    plt.ylabel("Score")
    plt.title("Rolling F1 and Accuracy Scores (Multiclass)")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
    print("Средна Macro F1:", df_results['F1_macro'].mean())
```

```python
from sklearn.metrics import accuracy_score

def classwise_accuracy(y_true, y_pred, class_label):
    mask = y_true == class_label
    if mask.sum() == 0:
        return np.nan
    return accuracy_score(y_true[mask], y_pred[mask])

acc_buy, acc_hold, acc_sell = [], [], []
start_indices = []

for start in range(0, len(X) - window_size - step, step):
    end = start + window_size
    X_train, y_train = X.iloc[start:end], y.iloc[start:end]
    X_test, y_test = X.iloc[end:end+step], y.iloc[end:end+step]

    if len(set(y_train)) < 3 or len(set(y_test)) < 2:
        continue

    X_resampled, y_resampled = hold_focused_sampling(X_train, y_train)
    weights = get_sample_weights(y_resampled)
    model.fit(X_resampled, y_resampled, sample_weight=weights)
    y_pred = model.predict(X_test)

    acc_buy.append(classwise_accuracy(y_test, y_pred, 2))   # BUY = 2
    acc_hold.append(classwise_accuracy(y_test, y_pred, 1))  # HOLD = 1
    acc_sell.append(classwise_accuracy(y_test, y_pred, 0))  # SELL = 0
    start_indices.append(start)

plt.figure(figsize=(10, 5))
plt.plot(start_indices, acc_buy, label='Accuracy BUY (2)', marker='o')
plt.plot(start_indices, acc_hold, label='Accuracy HOLD (1)', marker='x')
plt.plot(start_indices, acc_sell, label='Accuracy SELL (0)', marker='s')
plt.xlabel('Start Index')
plt.ylabel('Accuracy per class')
plt.title('Rolling Class-wise Accuracy (BUY, HOLD, SELL)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print("Средна Accuracy BUY:", np.nanmean(acc_buy))
print("Средна Accuracy HOLD:", np.nanmean(acc_hold))
print("Средна Accuracy SELL:", np.nanmean(acc_sell))
```
