# Preparing datasets

```python
merged_df['article_date_v2'] = pd.to_datetime(merged_df['article_date_v2'], errors='coerce')
df_gold_prices['Date'] = pd.to_datetime(df_gold_prices['Date'], errors='coerce')

merged_df['V2Tone_num'] = merged_df['V2Tone'].apply(lambda x: float(str(x).split(',')[0]) if isinstance(x, str) else None)
merged_df['has_gold_theme'] = merged_df['Themes'].str.contains('GOLD', na=False).astype(int)

daily_sentiment = merged_df.groupby('article_date_v2').agg(
    avg_tone=('V2Tone_num', 'mean'),
    max_tone=('V2Tone_num', 'max'),
    tone_std=('V2Tone_num', 'std'),
    article_count=('V2Tone_num', 'count'),
    has_gold_theme=('has_gold_theme', 'max')

).reset_index().rename(columns={'article_date_v2': 'Date'})

df_merged = pd.merge(df_gold_prices, daily_sentiment, on='Date', how='inner')

df_merged = df_merged.sort_values(by='Date')
df_merged['Price_t-1'] = df_merged['Price'].shift(1)
df_merged['Price_change_t-1'] = df_merged['Price'] - df_merged['Price_t-1']
df_merged['MA_3'] = df_merged['Price'].rolling(window=3).mean()
df_merged['MA_7'] = df_merged['Price'].rolling(window=7).mean()
df_merged['Volatility_3d'] = df_merged['Price'].rolling(window=3).std()
df_merged['avg_tone_ma3'] = df_merged['avg_tone'].rolling(window=3).mean()
df_merged['avg_tone_ma7'] = df_merged['avg_tone'].rolling(window=7).mean()

future_price = df_merged['Price'].shift(-3)
growth = (future_price - df_merged['Price']) / df_merged['Price']
df_merged['Target_up_0_3pct_3d'] = (growth > 0.003).astype(int)

df_merged = df_merged.dropna()
```

```python
features = [
    'Price', 'Open', 'High', 'Low', 'Change %',
    'avg_tone', 'max_tone', 'tone_std', 'article_count', 'has_gold_theme',
    'Price_t-1', 'Price_change_t-1', 'MA_3', 'MA_7', 'Volatility_3d',
    'avg_tone_ma3', 'avg_tone_ma7'
]

X = df_merged[features]
y = df_merged['Target_up_0_3pct_3d']
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import f1_score, accuracy_score
from imblearn.over_sampling import RandomOverSampler


results = []

window_size = 200
step = 50

for start in range(0, len(X) - window_size - step, step):
    end = start + window_size
    X_train = X.iloc[start:end]
    y_train = y.iloc[start:end]
    X_test = X.iloc[end:end+step]
    y_test = y.iloc[end:end+step]

    ros = RandomOverSampler(random_state=42)
    X_resampled, y_resampled = ros.fit_resample(X_train, y_train)

    model = XGBClassifier(
        n_estimators=150,
        max_depth=4,
        learning_rate=0.1,
        subsample=0.8,
        colsample_bytree=0.8,
        use_label_encoder=False,
        eval_metric='logloss'
    )
    model.fit(X_resampled, y_resampled)

    y_pred = model.predict(X_test)
    f1 = f1_score(y_test, y_pred, zero_division=0)
    acc = accuracy_score(y_test, y_pred)

    results.append({'Start': start, 'F1-score': f1, 'Accuracy': acc})
```

```python
import matplotlib.pyplot as plt

print("Средна Accuracy:", results['Accuracy'].mean())
print("Средна F1-score:", results['F1-score'].mean())


df_results = pd.DataFrame(results)
plt.figure(figsize=(10, 5))
plt.plot(df_results['Start'], df_results['F1-score'], label='F1-score', marker='o')
plt.plot(df_results['Start'], df_results['Accuracy'], label='Accuracy', marker='x')
plt.title('Rolling Evaluation (XGBoost)')
plt.xlabel('Start Index')
plt.ylabel('Score')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# GridSearch

```python
from xgboost import XGBClassifier
from sklearn.metrics import f1_score, accuracy_score
from imblearn.over_sampling import RandomOverSampler
from sklearn.model_selection import GridSearchCV

results = []

window_size = 200
step = 50

for start in range(0, len(X) - window_size - step, step):
    end = start + window_size
    X_train = X.iloc[start:end]
    y_train = y.iloc[start:end]
    X_test = X.iloc[end:end+step]
    y_test = y.iloc[end:end+step]

    # 💡 Балансирање на класите
    ros = RandomOverSampler(random_state=42)
    X_resampled, y_resampled = ros.fit_resample(X_train, y_train)

    # 🎯 GridSearch (Hyperparameter tuning)
    param_grid = {
        'max_depth': [3, 4, 5],
        'learning_rate': [0.05, 0.1, 0.2],
        'n_estimators': [100, 150, 200],
        'subsample': [0.7, 0.8, 1.0]
    }

    grid = GridSearchCV(
        estimator=XGBClassifier(use_label_encoder=False, eval_metric='logloss'),
        param_grid=param_grid,
        scoring='f1',
        cv=3,
        n_jobs=-1
    )
    grid.fit(X_resampled, y_resampled)
    best_model = grid.best_estimator_
    print("Best Params:", grid.best_params_)

    # 📊 Предвидување
    y_pred = best_model.predict(X_test)
    f1 = f1_score(y_test, y_pred, zero_division=0)
    acc = accuracy_score(y_test, y_pred)

    results.append({'Start': start, 'F1-score': f1, 'Accuracy': acc})
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df_results = pd.DataFrame(results)

print("Средна Accuracy:", df_results['Accuracy'].mean())
print("Средна F1-score:", df_results['F1-score'].mean())

plt.figure(figsize=(10, 5))
plt.plot(df_results['Start'], df_results['Accuracy'], label='Accuracy', marker='x')
plt.plot(df_results['Start'], df_results['F1-score'], label='F1-score', marker='o')
plt.xlabel("Start Index")
plt.ylabel("Score")
plt.title("Rolling Evaluation with Hyperparameter Tuning")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
